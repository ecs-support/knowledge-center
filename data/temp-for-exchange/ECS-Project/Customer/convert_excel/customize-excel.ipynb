# Customize Excel for section 29

```python
# import numpy and pandas
import numpy as np
import pandas as pd

# used for dates
import datetime
from datetime import datetime, date

print(f"pandas  version = {pd.__version__}")
pd.Timestamp.today()
```

```python
df = pd.read_excel('workbook.xlsx', 'Sheet1', index_col=None, na_values=['NA'],header=0,converters={'F':str,'H':str})
df.head()
```

```python
df['D'] = pd.to_datetime(df['D'], dayfirst = True)
df.head()
```

```python
df.info()
```

```python
# df['hour'] = pd.to_numeric(df['hour']) # แปลงเป็น Numeric
# df['hour'] = df['hour'].astype('int') # อีกวิธีในการแปลงค่า สามารถใช้วิธีนี้แปลงเป็น float ได้

df['F'] = df['F'].astype('str')
df.head()
```

```python
df['D'] = df['D'].dt.strftime('%d/%m/%Y')
df['E'] = df['E'].dt.strftime('%d/%m/%Y')
df['G'] = df['G'].dt.strftime('%d/%m/%Y')
df.head()
```

```python
df['A'].replace('-.', '', regex=True, inplace=True)

df.head()
```

```python
df.to_excel('data_1.xlsx')
```
