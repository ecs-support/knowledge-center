```python
# download from Google Sheet

# https://docs.google.com/spreadsheets/d/1dDgSxN2E2cVoLuOfyauDAuMlvx8nR6rWuBsFhcFFpSs/edit?usp=sharing

url='https://docs.google.com/spreadsheets/d/1dDgSxN2E2cVoLuOfyauDAuMlvx8nR6rWuBsFhcFFpSs/export?gid=0&format=csv'
df=pd.read_csv(url)
df.head(50)
```

```python
# Date & Time Manage

from datetime import date, timedelta

def get_first_day(dt, d_years=0, d_months=0):
    # d_years, d_months are "deltas" to apply to dt
    y, m = dt.year + d_years, dt.month + d_months
    a, m = divmod(m-1, 12)
    return date(y+a, m+1, 1)

def get_last_day(dt):
    return get_first_day(dt, 0, 1) + timedelta(-1)

def get_year_thai(d):
    
    '''แปลงข้อมูลเป็นปี พ.ศ.'''
    
    return d.replace(d.year+543)
```

```python
# ดึงข้อมูลวันที่ปัจจุบันจากเครื่อง
d = datetime.now()
d = d.replace(month=(d.month+1)) # เดือนถัดไป + 1 month
d= get_year_thai(d) # เปลี่ยนให้เป็น พ.ศ.

first_day = (get_first_day(d).strftime('%d''%m''%Y')) # หาวันแรกของเดือน
last_day = (get_last_day(d).strftime('%d''%m''%Y')) # หาวันสุดท้ายของเดือน
```

# จัดเรียงข้อมูลให้ตรงตาม format กรมศุลฯ

```python
ex = {}  # Create blank dict
im = {}
for i in range(len(df['Currency Code'])):
    ex[i] = (df['Currency Code'][i]),(first_day),(last_day),('%011.6f'%df['Export'][i]),('RFXEI02405256214423865A')
    im[i] = (df['Currency Code'][i]),(first_day),(last_day),('%011.6f'%df['Import'][i]),('RFXEI02405256214423865A')
    
# จัดรูปแบบข้อมูลเพื่อเตรียมเขียนลงไฟล์
## ขาออก
lst_ex = []
for i in ex.keys():
    e = ''.join(ex[i])
    lst_ex.append(e)    
data = '\n'.join(lst_ex) # กำหนดให้ขึ้นบรรทัดใหม่
# write data to text file
f = open('RFECU.txt','w',encoding='utf-8')
f.write (str(data))

# ขาเข้า
lst_im = []
for i in im.keys():
    e = ''.join(im[i])
    lst_im.append(e)
data = '\n'.join(lst_im)
# write data to text file
f = open('RFICU.txt','w',encoding='utf-8')
f.write (str(data))
f.close()
print('success')
```
