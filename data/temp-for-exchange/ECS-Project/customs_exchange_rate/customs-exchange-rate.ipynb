```python
import pandas as pd
import requests
import bs4
from bs4 import BeautifulSoup
import numpy as np

# used for dates
import datetime
from datetime import datetime, date

#from google.colab import files

url = 'http://www.customs.go.th/content_special.php?lang=en&link=exch_search.php&&left_menu=nmenu_esevice_003'
data = pd.read_html(url)  # First Page
df = data[2]

from datetime import date, timedelta

def get_first_day(dt, d_years=0, d_months=0):
    # d_years, d_months are "deltas" to apply to dt
    y, m = dt.year + d_years, dt.month + d_months
    a, m = divmod(m-1, 12)
    return date(y+a, m+1, 1)

def get_last_day(dt):
    return get_first_day(dt, 0, 1) + timedelta(-1)

def get_year_thai(d):
    
    '''แปลงข้อมูลเป็นปี พ.ศ.'''
    
    return d.replace(d.year+543)

# ดึงข้อมูลวันที่ปัจจุบันจากเครื่อง
d = datetime.now()
if d.month != 12:
    d = d.replace(month=(d.month+1)) # เดือนถัดไป + 1 month
    d = d.replace(d.year+543)# เปลี่ยนให้เป็น พ.ศ.
else:
    d = d.replace(month=1) # เดือนถัดไป + 1 month
    d=  d.replace(d.year+544) # เปลี่ยนให้เป็น พ.ศ.

first_day = (get_first_day(d).strftime('%d''%m''%Y')) # หาวันแรกของเดือน
last_day = (get_last_day(d).strftime('%d''%m''%Y')) # หาวันสุดท้ายของเดือน

ex = {}  # Create blank dict
im = {}
for i in range(len(df['Currency Code'])):
    ex[i] = (df['Currency Code'][i]),(first_day),(last_day),('%011.6f'%df['Export'][i]),('RFXEI02405256214423865A')
    im[i] = (df['Currency Code'][i]),(first_day),(last_day),('%011.6f'%df['Import'][i]),('RFXEI02405256214423865A')
    
# จัดรูปแบบข้อมูลเพื่อเตรียมเขียนลงไฟล์
## ขาออก
lst_ex = []
for i in ex.keys():
    e = ''.join(ex[i])
    lst_ex.append(e)    
data = '\n'.join(lst_ex) # กำหนดให้ขึ้นบรรทัดใหม่
# write data to text file
f = open('RFECU.txt','w',encoding='utf-8')
f.write (str(data))


# ขาเข้า
lst_im = []
for i in im.keys():
    e = ''.join(im[i])
    lst_im.append(e)
data = '\n'.join(lst_im)
# write data to text file
f = open('RFICU.txt','w',encoding='utf-8')
f.write (str(data))

f.close()

# excel
df.to_excel('exchange-rate.xlsx', sheet_name='Sheet1')

# Download เฉพาะกรณี run บน google drive
#files.download('RFECU.txt')
#files.download('RFICU.txt')
#files.download('exchange-rate.xlsx')

print('success')
```

```python
# กำหนดชื่อ Column ใหม่

n = ['ลำดับ','ประเทศ','สกุลเงิน','รหัสสกุลเงิน','การส่งออก','การนำเข้า']

df['Export'] = ('%.3f' %df['Export'])
df
```

```python
def create_md(df):
    #col_name = df.columns
    col_name = n
    md = "| "
    for i in range (len(col_name)):
        
        md += col_name[i]+" |"
    md += '\n'+"|"
    for i in range (len(col_name)):
        word = len(col_name[i])
        if word <= 5:
            md += "-----|"
        else:
            md += "----------|"
    md += '\n' 
  
    
    lst_heads = list(col_name)
    for col_idx,data in df.iterrows():    
      # md += "|"+ str(col_idx+1)+"|"
        md +=  "| "
        for i in range(len(lst_heads)):
          
            md += str(data[i]) + '|'
        md += '\n'    
    print(md)       
```
